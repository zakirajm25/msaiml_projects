```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# Create a sample dataset with customer annual income, spending score, and age
data = {
    'AnnualIncome': [
        15, 15.5, 16, 16.5, 17, 17.5, 18, 18.5, 19, 19.5, 
        20, 20.5, 21, 21.5, 22, 22.5, 23, 23.5, 24, 24.5, 
        25, 25.5, 26, 26.5, 27, 27.5, 28, 28.5, 29, 29.5, 
        30, 30.5, 31, 31.5, 32, 32.5, 33, 33.5, 34, 34.5, 
        35,   # Normal points
        80, 85, 90  # Outliers
    ],
    'SpendingScore': [
        39, 42, 45, 48, 51, 54, 57, 60, 63, 66,
        69, 72, 75, 78, 81, 84, 87, 90, 93, 96,
        6, 9, 12, 15, 18, 21, 24, 27, 30, 33,
        5, 8, 11, 14, 17, 20, 23, 26, 29, 32,
        56,   # Normal points
        2, 3, 100  # Outliers
    ],
    'Age': [
        20, 20.5, 21, 21.5, 22, 22.5, 23, 23.5, 24, 24.5, 
        25, 25.5, 26, 26.5, 27, 27.5, 28, 28.5, 29, 29.5, 
        30, 30.5, 31, 31.5, 32, 32.5, 33, 33.5, 34, 34.5, 
        35, 35.5, 36, 36.5, 37, 37.5, 38, 38.5, 39, 39.5, 
        40,   # Normal points
        15, 60, 70  # Outliers
    ]
}

df = pd.DataFrame(data)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Normalizing the data
scaler = StandardScaler()
scaled = scaler.fit_transform(df)

# Converting back to DataFrame for easy handling
df_scaled = pd.DataFrame(scaled, columns=['AnnualIncome', 'SpendingScore','Age'])

print(df_scaled.head())
```

```python
# Plotting the 3D Scatter Plot
fig = plt.figure(figsize = (10,7))
ax = fig.add_subplot(111, projection='3d')

# Scatter Plot
ax.scatter(df_scaled['AnnualIncome'], df_scaled['SpendingScore'], df_scaled['Age'], c='blue', marker = 'o')

# Adding labels
ax.set_xlabel('Annual Income (Scaled)')
ax.set_ylabel('Spending Score (Scaled)')
ax.set_zlabel('Age (Scaled)')

plt.show()
```

```python
# Applying PCA to reduce dimensions from 3 to 2
pca = PCA(n_components=2)
df_pca = pca.fit_transform(scaled)

# Converting back to DataFrame for easy handling
df_pca = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])
print(df_pca.head())
```

```python
# Plotting the PCA components
plt.scatter(df_pca['PCA1'], df_pca['PCA2'])
plt.title('PCA - Dimensionality Reduction')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
# PCA: merge original features with PCA coordinates
df_pca_result = df.copy()
df_pca_result['PCA1'] = df_pca['PCA1']
df_pca_result['PCA2'] = df_pca['PCA2']

# Flag outliers
df_pca_result['Outlier'] = np.where(
    (df_pca_result['AnnualIncome'] > 70) |
    (df_pca_result['SpendingScore'] > 90) |
    (df_pca_result['Age'] > 55) | (df_pca_result['Age'] < 18),
    'Yes', 'No'
)

# Show explained variance & loadings
print("PCA explained variance ratio:", np.round(pca.explained_variance_ratio_, 3))
pca_loadings = pd.DataFrame(
    pca.components_,
    columns=['AnnualIncome', 'SpendingScore', 'Age'],
    index=['PCA1', 'PCA2']
)
print("\nPCA loadings (feature contributions):")
print(pca_loadings)

# Scatterplot
plt.figure(figsize=(7,6))
sns.scatterplot(
    data=df_pca_result, x='PCA1', y='PCA2', hue='Outlier',
    palette={'Yes':'red','No':'blue'}, s=70, edgecolor='k'
)
plt.title('PCA: Outlier Highlight')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Outlier')
plt.show()

# Explained variance bar chart
plt.figure(figsize=(5,4))
plt.bar(['PCA1', 'PCA2'], pca.explained_variance_ratio_ * 100, color=['#4c72b0','#55a868'])
plt.ylabel('Variance Explained (%)')
plt.title('PCA Explained Variance')
for i, v in enumerate(pca.explained_variance_ratio_ * 100):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center')
plt.ylim(0, 100)
plt.show()

# Loadings heatmap
plt.figure(figsize=(6,3))
sns.heatmap(pca_loadings, annot=True, fmt=".2f", cmap='coolwarm', center=0,
            cbar_kws={'label': 'Feature Contribution'})
plt.title('PCA Loadings by Feature')
plt.show()
```

```python
# Applying t-SNE to reduce dimensions to 2
tsne = TSNE(n_components=2, perplexity=3, random_state=42)
df_tsne = tsne.fit_transform(scaled)

# Converting the t-SNE result back to a DataFrame
df_tsne = pd.DataFrame(df_tsne, columns=['t-SNE1', 't-SNE2'])
print(df_tsne.head())
```

```python
# Plotting the t-SNE components
plt.scatter(df_tsne['t-SNE1'], df_tsne['t-SNE2'])
plt.title('t-SNE - Dimensionality Reduction')
plt.xlabel('t-SNE1')
plt.ylabel('t-SNE2')
plt.show()
```

```python
# t-SNE: merge original features with t-SNE coordinates
df_tsne_result = df.copy()
df_tsne_result['t-SNE1'] = df_tsne['t-SNE1']
df_tsne_result['t-SNE2'] = df_tsne['t-SNE2']

# Flag outliers
df_tsne_result['Outlier'] = np.where(
    (df_tsne_result['AnnualIncome'] > 70) |
    (df_tsne_result['SpendingScore'] > 90) |
    (df_tsne_result['Age'] > 55) | (df_tsne_result['Age'] < 18),
    'Yes', 'No'
)

# Scatterplot with outlier highlight
plt.figure(figsize=(7,6))
sns.scatterplot(
    data=df_tsne_result, x='t-SNE1', y='t-SNE2', hue='Outlier',
    palette={'Yes':'red','No':'blue'}, s=70, edgecolor='k'
)
plt.title('t-SNE: Outlier Highlight')
plt.xlabel('t-SNE1')
plt.ylabel('t-SNE2')
plt.legend(title='Outlier')
plt.show()
```
